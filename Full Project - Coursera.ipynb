# Task 1 – Import required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import random
import matplotlib.pyplot as plt
```

# Task 2 – Generate random data for the social media data

```python


categories=["Food", "Travel", "Fashion", "Fitness", "Music", "Culture", "Family", "Health"]
n=500

data={
    'Date': pd.date_range('2021-01-01', periods=n),
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes' :np.random.randint(0,10000,size=n)

}
df=pd.DataFrame(data)
df.head()
```

# Task 3 – Load the data into a Pandas DataFrame and Explore the data

```python
print("\nDataFrame Information:")
df.info()

print("\nDataFrame Description:")
df.describe()

category_counts = df['Category'].value_counts()
print("\nCount of each 'Category' element:")
print(category_counts)
```

# Task 4 – Clean the data

```python
nan_count = df.isnull().sum()
print('Number of NaN values:', nan_count)
```

```python
df.dropna(inplace = True)
df.drop_duplicates(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df['Likes']=pd.to_numeric(df['Likes'])
print(df.to_string()) 
df.dtypes
```

# Task 5– Visualize and Analyze the data

```python
sns.histplot(data=df,x="Likes")
plt.title('Histogram of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x=df["Category"],y=df["Likes"], fill=False)
plt.title('Boxplot of Likes by Category')
plt.show()
```

```python
print(f"Mean Likes: {df['Likes'].mean()}")
```

```python
print(f"\nMean Likes by Category:{df.groupby('Category')['Likes'].mean()}")
```
